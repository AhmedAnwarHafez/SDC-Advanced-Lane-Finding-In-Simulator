In Udacity's Self-Driving Cars Nanodegree, we correct this distortion for real cameras. I'll repeat the same exact same steps but for the [Car Simulator](https://github.com/udacity/self-driving-car-sim).

## Step 2: Camera Caliberation 

Note: I took around 27 caliberation images and you can find them under caliberation-images directory.
There is no need to follow the same steps in the section above, you can simply download the images and start from there.

The next step to do is to find the corners of the chessboard and visualize to make sure that everything is in the right place.

```python
from functions import *
import pickle
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
from ipywidgets import FloatSlider, interactive
from moviepy.editor import VideoFileClip

%matplotlib inline
```

The chessboard image I used have 9 horizontal and 6 vertical corners.

#### Here is an example before and after finding the corners

```python
nx = 9
ny = 6
objpoints = []
imgpoints = []

objp = np.zeros((nx*ny,3), np.float32)
objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)

raw = cv2.imread("calibration-images/test (26).png")
gray = cv2.cvtColor(raw, cv2.COLOR_BGR2GRAY)
draw_image = raw.copy()

# Find the chessboard corners
ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)

# If found, draw corners
if ret == True:
    imgpoints.append(corners)
    objpoints.append(objp)
    cv2.drawChessboardCorners(draw_image,  (nx, ny), corners, ret)

    f, axarr = plt.subplots(1, 2, figsize = (20,10))
    axarr[0].imshow(raw)
    axarr[0].set_title("before", fontsize=30)
    axarr[1].imshow(draw_image)
    axarr[1].set_title("after", fontsize=30)
```

The next thing to do is compute the camera matrix and the distortion coeffients

```python
images = glob.glob('calibration-images/test (*).png')

dist, mtx = compute_dist(images)
 
dist_pickle = {}
dist_pickle["mtx"] = mtx
dist_pickle["dist"] = dist
pickle.dump( dist_pickle, open( "calibration-images/camera_dist_pickle.p", "wb" ) )

print("pickle file saved")
```

OK, here is an example of an image before and after the distortion correction.

As you can see that there is a slight distortion happening and there is a thin black area at the left of the image.

```python
with open('calibration-images/camera_dist_pickle.p', 'rb') as fp:
    camera_dict = pickle.load(fp)
    
img = cv2.imread("calibration-images/test.png")

dist = camera_dict['dist']
mtx = camera_dict['mtx']


dst = cv2.undistort(img, mtx, dist, None, mtx)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(img)
ax1.set_title('Original Image', fontsize=30)
ax2.imshow(dst)
ax2.set_title('Undistorted Image', fontsize=30)
```

## Step 4: Warping Perspective

I used the `FloatSlider` from `ipywidgets` module to tweak the `src` vertices and after a few experiments with `src` and `destination` matrices, I think this is the best top-down view that I settled on.

The output shows that the lane lines are aligned with the yellow rectangle.

```python
def update(leftx, lefty, rightx, righty):
    img = plt.imread('test-images/straight5.jpg')
    img = img[0:140,:,:]
    undist = cv2.undistort(img, mtx, dist, None, mtx)
    undist_graph = undist.copy()

    height_image = img.shape[0]
    width_image  = img.shape[1]

    rectangle = np.float32([[250, 0],
                      [80, 0],
                      [80, height_image], 
                      [250, height_image]])

    top_left = (129, 80)
    top_right = (210, 80)
    bottom_left = (-0, int(height_image * 0.93))
    bottom_right =  (width_image+0, int(height_image * 0.93))

    src = np.float32([[top_right[0], top_right[1]],
                      [top_left[0], top_left[1]],
                      [bottom_left[0], bottom_left[1]], 
                      [bottom_right[0], bottom_right[1]]])


    offset = 50

    dst = np.float32([[width_image - offset, 0 ], #top right
                    [offset, 0], # top left
                    [offset, height_image], # bottom left
                    [width_image- offset, height_image]]) #bottom right

    cv2.polylines(undist_graph, np.int32([src]), 1, (255,0,0), thickness=2)
    cv2.polylines(undist_graph, np.int32([dst]), 1, (255,255,0), thickness=2)

    M = cv2.getPerspectiveTransform(src, dst)

    top_down = cv2.warpPerspective(undist, M, (width_image, height_image), flags=cv2.INTER_LINEAR)

    cv2.polylines(top_down, np.int32([rectangle]), 1, (255,255,0), thickness=2)

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    f.tight_layout()
    ax1.imshow(undist_graph)
    ax1.set_title('Original Image', fontsize=20)
    ax2.imshow(top_down)
    ax2.set_title('Undistorted and Warped Image', fontsize=20)
    plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

leftx = FloatSlider(min=0.0, max=500, step=1, value=120)
lefty = FloatSlider(min=0.0, max=500.0, step=1, value=80)
rightx = FloatSlider(min=0.0, max=500.0, step=1, value=215)
righty = FloatSlider(min=0.0, max=500.0, step=1, value=80)

interactive(update,
            leftx = leftx,
            lefty = lefty,
            rightx= rightx,
            righty= righty )
```

## Step 2: Finding the Lines

```python
# the offset of the destination matrix
offset = 100
    
def process_image(raw):
    zero = np.zeros_like(raw).astype(np.uint8)
    diag1 = zero 
    diag2 = zero
    diag3 = zero
    diag4 = zero
    diag5 = zero
    diag6 = zero
    diag8 = zero
    diag9 = zero
    diag10 = zero 
    diag11 = zero
    curvature_meters = 0
    off_center = 0
    
    undist = cv2.undistort(raw, mtx, dist, None, mtx)

    preprocessImage = np.zeros_like(undist[:,:,0])

    gradx = abs_sobel_thresh(undist, orient='x', thresh=(20, 255))
    grady = abs_sobel_thresh(undist, orient='y', thresh=(30, 255))
    c_binary = color_threshold(undist, sthresh=(70, 255), rthresh=(200, 255))
    preprocessImage[((gradx == 1) & (grady == 1) & (c_binary == 1))] = 255
    #preprocessImage[(c_binary == 1)] = 255
    perspective_result = draw_perpestive_boundaries(undist, offset=offset)
    undist_warped, _ = warp(undist, offset=offset)
    binary_warped, Minv = warp(preprocessImage, offset)
    result = np.dstack((binary_warped,binary_warped,binary_warped)) * 255
    
    left_fit, right_fit, out_img = blind_search(binary_warped)
    
    search_result = draw_lanes(binary_warped, left_fit, right_fit)
    left_curverad, right_curverad = compute_curvature(binary_warped, left_fit, right_fit)
    result = map_lane(Minv,undist,binary_warped, left_fit, right_fit)


    diag1 = result 
    diag2 = np.dstack((gradx, gradx, gradx))*255
    diag3 = np.dstack((grady, grady, grady))*255
    diag4 = np.dstack((c_binary, c_binary, c_binary))*255
    diag5 = np.dstack((preprocessImage, preprocessImage, preprocessImage))
    diag6 = search_result
    diag8 = raw
    diag9 = perspective_result
    diag10 = undist_warped 
    diag11 = out_img
        
    return plot_mosaic(None, None, diag1, diag2, diag3, diag4, diag5, diag6, diag8, diag9, diag10, diag11)
```

```python
f, axarr = plt.subplots(2, 1, figsize=(40,20))

#images = glob.glob('./test-images/straight*.jpg')
images = ['./test-images/straight5.jpg','./test-images/straight4.jpg', ]

for idx, fname in enumerate(images):
    img = plt.imread(fname)
    
    img = img[0:140,:,:]
    result = process_image(img)

    axarr[idx].imshow(result)
```
